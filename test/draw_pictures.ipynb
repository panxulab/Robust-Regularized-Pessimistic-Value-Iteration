```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 5, 10])
PEVI= np.array([0.803972403,	4.066360394,	8.153016567
])
DRPVI = np.array([1.701996009,	6.773483276,	13.39591948

])
DRVI_L = np.array([1.64305981,	7.012542407,	15.04436302
])

R2PVI_TV = np.array([1.078054269,	5.04795448,	8.992568334
])
R2PVI_KL = np.array([1.04103756,	5.018609206,	9.031555414
])
width = 0.5


x1 = x - 2 * width
x2 = x - width
x3 = x
x4= x + width
x5 = x + 2 * width

colors_blue = plt.cm.Blues(np.linspace(0.2, 0.8, 5))
colors_orange = plt.cm.Oranges(np.linspace(0.2, 0.8, 5))

plt.bar(x1, PEVI, width=width, label='PEVI (base)', color='gray')
plt.bar(x2, R2PVI_TV, width=width, label='R2PVI-TV (ours)', color=colors_blue[2])
plt.bar(x3, R2PVI_KL, width=width, label='R2PVI-KL (ours)', color=colors_blue[3])
plt.bar(x4, DRPVI, width=width, label='DRPVI', color=colors_orange[2])
plt.bar(x5, DRVI_L, width=width, label='DRVI-L', color=colors_orange[3])

#plt.bar(x_d201, time_d201, width=width, label=r'$SDRVI-chi2$', color='lightseagreen')

plt.xlabel(r'Sample Size N (x$10^3$)', size=16)
plt.ylabel('Time(s)', size=16)
plt.xticks(x, x)
plt.grid(alpha=0.3)
plt.legend(fontsize=16)
plt.savefig(f'time_N.pdf', dpi=1000, bbox_inches='tight', pad_inches=0.0)
```
